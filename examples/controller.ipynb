# Controller Testing with Sarsa

```python
from velora import RLController
import torch

from velora.agent import Sarsa, ExpectedSarsa, QLearning
from velora.env.gym import GymEnvHandler
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
RUN_NAME = "test1"
```

```python
sarsa = RLController(
    config_filepath="config/sarsa.yaml",
    env_type=GymEnvHandler,
    agent_type=Sarsa,
    device=device,
)

q_learning = RLController(
    config_filepath="config/sarsa.yaml",
    env_type=GymEnvHandler,
    agent_type=QLearning,
    device=device,
)

expected_sarsa = RLController(
    config_filepath="config/sarsa.yaml",
    env_type=GymEnvHandler,
    agent_type=ExpectedSarsa,
    device=device,
)
```

```python
sarsa.train(RUN_NAME)
```
