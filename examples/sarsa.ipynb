# Sarsa

```python
import gymnasium as gym
import torch

from velora.models import Sarsa, ExpectedSarsa, QLearning
from velora.utils import load_config, plot_state_values
```

```python
config = load_config("config/sarsa.yaml")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
from velora.utils import ignore_empty_dicts

ignore_empty_dicts(config.model_dump(exclude='model', exclude_none=True))
```

```python
run_name = "test3"
```

```python
agent = Sarsa(
    config=config,
    env=gym.make(config.env.name),
    device=device,
)

Q = agent.train(run_name)
```

```python
plot_state_values(Q.as_state_values(), (4, 12))
```

# Q-Learing

```python
agent = QLearning(
    config=config,
    env=gym.make(config.env.name),
    device=device,
)

Q = agent.train(run_name)
```

```python
plot_state_values(Q.as_state_values(), (4, 12))
```

```python
agent = ExpectedSarsa(
    config=config,
    env=gym.make(config.env.name),
    device=device,
)

Q = agent.train(run_name)
```

```python
plot_state_values(Q.as_state_values(), (4, 12))
```
